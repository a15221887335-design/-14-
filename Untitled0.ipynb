```python
# main.py
import data_manager  # 引入皓翔的檔案
import features      # 引入宣文的檔案

def print_table(task_list):
    """美化輸出表格的函式"""
    print("\n--------------------------------------------------------------")
    print(f"{'編號':<6}{'狀態':<6}{'優先度':<8}{'提醒日期/時間':<20}{'事項名稱'}")
    print("--------------------------------------------------------------")
    for t in task_list:
        status = "☑" if t["completed"] else "☐"
        p_label = "🌱低"
        if t["priority"] == 1: p_label = "🔥高"
        if t["priority"] == 2: p_label = "⚡中"
        print(f"#{t['id']:<5}{status:<7}{p_label:<8}{t['date']} {t['time']:<7}{t['title']}")
    print("--------------------------------------------------------------")

def main_menu():
    # 程式一啟動，先叫皓翔去讀取文字檔舊資料
    data_manager.load_from_txt()

    while True:
        print("\n===== 📝 備忘錄管理系統 =====")
        print("1. 查看目前清單 (預設建立順序)")
        print("2. 查看目前清單 (依優先度 高→低)")
        print("3. 查看目前清單 (依提醒時間 近→遠)")
        print("4. 新增待辦事項")
        print("5. 修正/編輯事項")
        print("6. 打勾完成 / 取消完成")
        print("7. 刪除待辦事項")
        print("8. 查看歷史紀錄與關鍵字搜尋")
        print("0. 結束程式")
        print("=============================")

        choice = input("👉 請輸入功能編號 (0-8): ").strip()

        if choice == "1":
            print_table(data_manager.tasks)

        elif choice == "2":
            # 把皓翔的資料，拿給宣文去排優先度
            sorted_tasks = features.sort_tasks(data_manager.tasks, sort_by="priority")
            print_table(sorted_tasks)

        elif choice == "3":
            # 把皓翔的資料，拿給宣文去排時間
            sorted_tasks = features.sort_tasks(data_manager.tasks, sort_by="time")
            print_table(sorted_tasks)

        elif choice == "4":
            title = input("輸入事項名稱: ")
            date_str = input("輸入日期 (如 2026-05-24): ")
            time_str = input("輸入時間 (如 14:30): ")
            priority = input("輸入優先度 (1:高, 2:中, 3:低): ")

            # 防呆檢查
            if priority not in ["1", "2", "3"]:
                print("❌ 優先度只能填 1, 2, 3，請重新操作！")
                continue

            data_manager.add_task(title, date_str, time_str, priority)
            print("✅ 新增成功！")

        elif choice == "5":
            print_table(data_manager.tasks)
            tid = int(input("請輸入要【修正】的事項編號: "))
            title = input("輸入新名稱: ")
            date_str = input("輸入新日期: ")
            time_str = input("輸入新時間: ")
            priority = int(input("輸入新優先度 (1-3): "))

            if data_manager.edit_task(tid, title, date_str, time_str, priority):
                print("📝 修改成功！")
            else:
                print("❌ 找不到該編號。")

        elif choice == "6":
            print_table(data_manager.tasks)
            tid = int(input("請輸入要【打勾/切換】的事項編號: "))
            if data_manager.toggle_task(tid):
                print("🔔 狀態切換成功！")
            else:
                print("❌ 找不到該編號。")

        elif choice == "7":
            print_table(data_manager.tasks)
            tid = int(input("請輸入要【刪除】的事項編號: "))
            deleted = data_manager.delete_task(tid)
            if deleted:
                print(f"🗑️ 已刪除事項：【{deleted['title']}】")
            else:
                print("❌ 找不到該編號。")

        elif choice == "8":
            print("\n📜 【歷史紀錄專區（所有已打勾事項）】")
            completed_tasks = []
            for t in data_manager.tasks:
                if t["completed"] == True:
                    completed_tasks.append(t)
            print_table(completed_tasks)

            want_search = input("🔍 是否要進行關鍵字搜尋？(y/n): ").strip().lower()
            if want_search == "y":
                kw = input("請輸入關鍵字: ")
                # 叫宣文去幫忙搜關鍵字
                result = features.search_tasks(data_manager.tasks, kw)
                print(f"\n🎯 包含「{kw}」的搜尋結果：")
                print_table(result)

        elif choice == "0":
            print("👋 系統已成功關閉！")
            break
        else:
            print("❌ 輸入錯誤，請輸入 0 到 8 之間的數字。")

if __name__ == "__main__":
    main_menu()
```
