```python
# 初始化範例資料（每一筆待辦事項都是一個 dict，全部存進 list 裡）
# 優先度用數字表示：1代表示「高」、2代表「中」、3代表「低」
tasks = [
    {"id": 1, "title": "準備期末專題簡報", "date": "2026-06-01", "time": "10:00", "priority": 1, "completed": False},
    {"id": 2, "title": "去健身房運動 1 小時", "date": "2026-05-25", "time": "19:30", "priority": 3, "completed": False},
    {"id": 3, "title": "繳交本月手機帳單", "date": "2026-05-24", "time": "12:00", "priority": 2, "completed": True}
]

# 用來記錄下一個新增的事項編號
id_counter = 4

# 用來存放「被刪除事項」的歷史紀錄清單
history_records = []


# === 功能 1：顯示清單與排序 ===
def display_tasks(sort_by="id"):
    if len(tasks) == 0:
        print("\n--- 目前沒有任何待辦事項 ---")
        return

    # 處理排序邏輯 (利用簡單的冒泡排序法原理或自訂排序)
    # 為了不破壞原本的 tasks 順序，複製一份出來排序
    display_list = list(tasks)

    # 泡沫排序法 (Bubble Sort)：最基礎、最好解釋的排序語法
    for i in range(len(display_list)):
        for j in range(0, len(display_list) - i - 1):

            if sort_by == "priority":
                # 依優先度排序：數字小的（1:高）排前面
                if display_list[j]["priority"] > display_list[j+1]["priority"]:
                    # 交換位置
                    display_list[j], display_list[j+1] = display_list[j+1], display_list[j]

            elif sort_by == "time":
                # 依日期時間排序：字串直接比大小（例如 "2026-05-24" < "2026-06-01"）
                datetime_j = display_list[j]["date"] + " " + display_list[j]["time"]
                datetime_j1 = display_list[j+1]["date"] + " " + display_list[j+1]["time"]
                if datetime_j > datetime_j1:
                    display_list[j], display_list[j+1] = display_list[j+1], display_list[j]

    # 印出表格外觀
    print("\n--------------------------------------------------------------")
    print(f"{'編號':<6}{'狀態':<6}{'優先度':<8}{'提醒時間':<18}{'事項名稱'}")
    print("--------------------------------------------------------------")
    for t in display_list:
        # 視覺回饋：根據完成狀態顯示不同符號
        status = "☑" if t["completed"] else "☐"

        # 轉換優先度顯示
        p_label = "🌱低"
        if t["priority"] == 1: p_label = "🔥高"
        if t["priority"] == 2: p_label = "⚡中"

        print(f"#{t['id']:<5}{status:<7}{p_label:<8}{t['date']} {t['time']:<7}{t['title']}")
    print("--------------------------------------------------------------")


# === 功能 2：新增事項 ===
def add_task(title, date_str, time_str, priority):
    global id_counter
    # 建立一個新的 dict
    new_task = {
        "id": id_counter,
        "title": title,
        "date": date_str,
        "time": time_str,
        "priority": int(priority),
        "completed": False
    }
    tasks.append(new_task)
    id_counter += 1
    print(f"✅ 成功新增事項：【{title}】")


# === 功能 3：修改/修正事項 ===
def edit_task(task_id, new_title, new_date, new_time, new_priority):
    for t in tasks:
        if t["id"] == task_id:
            t["title"] = new_title
            t["date"] = new_date
            t["time"] = new_time
            t["priority"] = int(new_priority)
            print(f"📝 成功修改編號 #{task_id} 的事項內容。")
            return
    print("❌ 找不到該編號的事項。")


# === 功能 4：打勾/切換完成狀態 ===
def toggle_task(task_id):
    for t in tasks:
        if t["id"] == task_id:
            # 簡單的布林值反轉 (! 邏輯)
            t["completed"] = not t["completed"]
            status = "☑ 已完成" if t["completed"] else "☐ 未完成"
            print(f"🔔 編號 #{task_id} 狀態已變更為：{status}")
            return
    print("❌ 找不到該編號的事項。")


# === 功能 5：刪除事項並移入歷史 ===
def delete_task(task_id):
    for t in tasks:
        if t["id"] == task_id:
            # 在刪除前，先複製一份放到歷史紀錄清單
            record = {
                "title": t["title"],
                "final_status": "已完成" if t["completed"] else "未完成（遭刪除）"
            }
            history_records.append(record)

            # 從進行中清單移除
            tasks.remove(t)
            print(f"🗑️ 已刪除事項：【{t['title']}】，此紀錄已移至歷史專區。")
            return
    print("❌ 找不到該編號的事項。")


# === 功能 6：查看歷史紀錄 ===
def display_history():
    print("\n📜 【所有歷史紀錄查看】")
    print("--------------------------------------------------------------")

    # 1. 撈出目前清單中已經打勾完成的事項
    for t in tasks:
        if t["completed"]:
            print(f"[狀態: ☑ 已完成] -> {t['title']}")

    # 2. 撈出已經被刪除的事項
    for record in history_records:
        print(f"[狀態: {record['final_status']}] -> {record['title']}")

    print("--------------------------------------------------------------")


# === 主程式迴圈（互動介面） ===
while True:
    print("\n===== 簡易備忘錄系統 =====")
    print("1. 查看待辦清單 (預設建立順序)")
    print("2. 查看待辦清單 (依優先度高到低)")
    print("3. 查看待辦清單 (依時間遠近)")
    print("4. 新增待辦事項")
    print("5. 修正/編輯事項內容")
    print("6. 打勾完成 / 取消完成")
    print("7. 刪除待辦事項")
    print("8. 查看所有歷史紀錄")
    print("0. 離開系統")

    choice = input("👉 請選擇操作項目 (0-8): ").strip()

    if choice == "1":
        display_tasks(sort_by="id")
    elif choice == "2":
        display_tasks(sort_by="priority")
    elif choice == "3":
        display_tasks(sort_by="time")

    elif choice == "4":
        title = input("輸入事項名稱: ")
        date_str = input("輸入日期 (如 2026-05-24): ")
        time_str = input("輸入時間 (如 14:30): ")
        priority = input("輸入優先度 (1:高, 2:中, 3:低): ")
        add_task(title, date_str, time_str, priority)

    elif choice == "5":
        display_tasks()
        tid = int(input("請輸入要【修正】的事項編號: "))
        title = input("輸入新名稱: ")
        date_str = input("輸入新日期: ")
        time_str = input("輸入新時間: ")
        priority = input("輸入新優先度 (1:高, 2:中, 3:低): ")
        edit_task(tid, title, date_str, time_str, priority)

    elif choice == "6":
        display_tasks()
        tid = int(input("請輸入要【打勾/切換】的事項編號: "))
        toggle_task(tid)

    elif choice == "7":
        display_tasks()
        tid = int(input("請輸入要【刪除】的事項編號: "))
        delete_task(tid)

    elif choice == "8":
        display_history()

    elif choice == "0":
        print("👋 系統已關閉，祝你專題順利！")
        break
    else:
        print("❌ 輸入錯誤，請重新輸入 0~8 的數字。")
```
