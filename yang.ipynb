```python
# data_manager.py

# 記憶體中的核心資料結構：存放多個字典的列表
tasks = []
id_counter = 1

def load_from_txt():
    """從 txt 檔案讀取資料"""
    global id_counter, tasks
    tasks.clear()

    try:
        # 開啟純文字檔讀取資料
        with open("todo_data.txt", mode="r", encoding="utf-8") as f:
            for line in f:
                line = line.strip() # 去除換行符號
                if not line:
                    continue

                # 用逗號切開文字，還原成欄位
                parts = line.split(",")
                task = {
                    "id": int(parts[0]),
                    "title": parts[1],
                    "date": parts[2],
                    "time": parts[3],
                    "priority": int(parts[part_index := 4]),
                    "completed": parts[5] == "True" # 字串轉布林值
                }
                tasks.append(task)
                if task["id"] >= id_counter:
                    id_counter = task["id"] + 1
    except:
        # 如果檔案不存在，就忽略錯誤（等下第一次存檔會自動建立）
        pass

def save_to_txt():
    """將目前的資料全部寫入 txt 檔案"""
    with open("todo_data.txt", mode="w", encoding="utf-8") as f:
        for t in tasks:
            # 把字典組合包裝成一列文字寫入
            line = f"{t['id']},{t['title']},{t['date']},{t['time']},{t['priority']},{t['completed']}\n"
            f.write(line)

def add_task(title, date_str, time_str, priority):
    """【新增】待辦事項"""
    global id_counter
    new_task = {
        "id": id_counter,
        "title": title,
        "date": date_str,
        "time": time_str,
        "priority": int(priority),
        "completed": False
    }
    tasks.append(new_task)
    id_counter += 1
    save_to_txt()

def edit_task(task_id, new_title, new_date, new_time, new_priority):
    """【修正】待辦事項"""
    for t in tasks:
        if t["id"] == task_id:
            t["title"] = new_title
            t["date"] = new_date
            t["time"] = new_time
            t["priority"] = int(new_priority)
            save_to_txt()
            return True
    return False

def toggle_task(task_id):
    """【打勾】切換完成狀態"""
    for t in tasks:
        if t["id"] == task_id:
            t["completed"] = not t["completed"]
            save_to_txt()
            return True
    return False

def delete_task(task_id):
    """【刪除】待辦事項"""
    for t in tasks:
        if t["id"] == task_id:
            tasks.remove(t)
            save_to_txt()
            return t
    return None
```
